```python
# module 11
# we will learn how to read and write data to excel files, json file and also sql database.
# in this series so far, we have been doing our work by reading the data from a csv file.
# but in data science, there are so many format where our data can be stored.
# by the end of this module, we will be able to get data to and from pandas no matter what data format we are using.
# now , if we want to read and write data from a specific file format,
# then we can see the time stamp to where we read and write each different format.
```

```python
# first look at csv files since we have already been using this throughout the series.
import pandas as pd

df=pd.read_csv("stackoverflow/survey_results_public.csv",index_col="Respondent")
schema_df=pd.read_csv("stackoverflow/survey_results_schema.csv",index_col="Column")

pd.set_option("display.max_columns", 85)
pd.set_option("display.max_rows", 85)

df.head()
```

```python
# here we can use read_csv() method to read csv files.
# and we have to enter the location of the file as an arguemnet.
# we can also use extra arguements in our read_csv() method.
# if we use the read_csv() method, we can see that this just create a dataframe with the data of our csv file.
```

```python
# now lets learn how to write this data back to csv.
# maybe we have to do an analysis on that data and have to save it so that we could work with that later.
# for example, lets we create a filter based on our country.
filt=(df["Country"]=="Bangladesh")
bd_df=df.loc[filt]
bd_df.head()
```

```python
# now lets say we want to export this filter to a csv file.
# to do this we can use the to_csv() method.
bd_df.to_csv("11. created_files/bd_modified.csv")
# we have to pass the location where we want to save our csv file as an arguement.
```

```python
# now if we go to that location we can se ethat we have a bd_modified.csv file there.
# now lets look at how to read and write some other formats.
# one format that we might run into is a tab limited file.
# these are the same thing as csv file.
# but instead of separated by commas, they are separated by tabs.
# if we want to read and write those files in pandas,
# we can still use the csn methods in it.
# but we need to pass in a custom separator.
# we can save a tab limited file by changing the extension ".tsv"
bd_df.to_csv("11. created_files/bd_modified.tsv", sep="\t")
# and we need to tell pandas about the separatior using sep arguement.
# files can be separated by anything (:,:-_) we have to just tell pandas about it.
```

```python
# now we can see that we have bd_modified.tsv
# and if we are reading in the tsv file, then we need to add the separator "\t" as the sep arguement
bd_modified2=pd.read_csv("11. created_files/bd_modified.tsv", sep="\t",index_col="Respondent")
bd_modified2.head()
```

```python
# now lets look at excel file format.
# it is the most popular format for working with this type of data.
# now if we want to writein excel then we need pip to install couple of packages.
# first one is "pip install xlwt" xlwt is the right older of .xlsx format.
# but if we want to write in .xlsx format, then we need to install "pip install xlwt openpyxl"
# if we want to read excel file then we need to install xlrd package "pip install xlwt openpyxl xlrd"
```

```python
# now lets write to an excel file, we can do that by using to_excel() method.
bd_df.to_excel("11. created_files/bd_modified.xlsx")
```

```python
# there are more advanced things that we can do with excel as well.
# excel works with different sheets.
# if we want to write over specific sheets then we can pass in a sheet arguement to theto_excel() method.
# and there is also way to start with different rows and columns as well.
# if we google this method name(to_excel()) then we can see all the arguements in the documentation.
```

```python
# now lets read the excel file.
# by default, it is going to load in the data with a default index.
# so we have to specify that we want our index column to be that column.
# to do this we can use read_excel() method.
test=pd.read_excel("11. created_files/bd_modified.xlsx", index_col="Respondent")
test.head()
```

```python
# json is also very popular for this kind of data.
# first lets write our modified dataframe to json file.
# we can use the to_json() method.
bd_df.to_json("11. created_files/bd_modified.json")
# we need to use the .json extension.
```

```python
# if we look at our json file we can see that this pretty like dictionary.
# there also different ways to write json files.
# lets say we want to this json list like instead of dictionary like.
# to do this we can change the orient arguement.
bd_df.to_json("11. created_files/bd_modified2.json", orient="records", lines=True)
# here "records" is just list type.
# and other arguement lines which we have setted to True, 
# it is just doing the list separated line by line which will make the data easier to read.
# if we want to know more about the arguemnets that we can use in to_json() method then we can google it.
```

```python
# now we can see that we have one response at a time.
# now lets read our json file.
# since we have used the arguements when we created the json file.
# we need to use that arguements when we read the data.
# we can read the json file using read_json() method.
json_test=pd.read_json("11. created_files/bd_modified2.json", orient="records", lines=True)
json_test.head()

# so if we are reading in json files and facing issues, then we might need to play around with some of those arguements
# so that our data is fit with those arguemnets.
```

```python
# lets learn how we can read and write data from SQL database.
# this is probably the most complicated one.
# because we need to have the database setup and all thta good stuffs.

# first lets say how we can connect with that database.
# we are going to need to install a package to do this.
# to connect to our database,  we need to "pip install SQLAlchemy"
# this is a very popularr ORM(Object Relational Mapper) for python that allows us to work more easily with databases.
# ORM is just a way to use python object to a database.
# depending in what database we are using we might not need to do anything here.
# but since we are using postgresql database here, 
# we also need to install psycopg2 package here which allows us to work with postgresql database.
# "pip install psycopg2-binary"
```

```python
# now lets import all the packages we need and connect to our postgresql database.
from sqlalchemy import create_engine
# this will allow us to connect to our database.

import psycopg2

# to create the engine we need to use the create_engine() function.
engine= create_engine("postgresql://postgres:as8as8as8as8@localhost:5432/as8_db")
# in the create_engine() function we need to pass the "postgres connection string" as an arguement.
# we need to pass the user name and password for our database.

# another thing is that we should put credentials(username/password) in the code like this.
# we always should use environment variable and config file to hide this information.
# so if we are using this in production code, using environment variable is recommanded to hide our password to be exposed.
# after the username and password we need to pass the host name and the port and finally the name of our database.

# now we can get a connection to that database.
# now just write our modified dataframe to a table of this database.
# this table doesn't need to currently exist.
# by default, it will create the table for us.
# if it does exist then we need to pass another arguement to handle that.
```

```python
# to write into that table we need to use the to_sql() function
bd_df.to_sql("Bangladesh",engine)
# here we have to pass the table name as the first arguement.
# and our database connection which we are calling "engine" as the second arguement.
```

```python
# now if we go to pgadmine we can see our table there.
# but what if we wanted to update our data and rewrite that to our database?

# if we try to run the same line again-
# bd_df.to_sql("Bangladesh",engine)
# we are going to get an error because this table already exists.
# if we want to write over a table, then we can add in additional arguement.
# and the arguement is "if_exists" and set that equal to what we want to do if this alrady exits.
# in our case that will be equal to "replace"
# but there are lots more option we could do here.
# we could throw an error(default), we can append the data and so on.
# if we are doing a daily script, when we are just appending data then we can just append the daily data to our existing table.
bd_df.to_sql("bangla",engine,if_exists="replace")
```

```python
# now that we write our data to the sql database table, lets see how to read in that data.
# we can do this by read_sql method.
bd_db=pd.read_sql("bangla",engine,index_col="Respondent")
# here we have to pass the table name as the first arguement.
# and our database connection which we are calling "engine" as the second arguement.
# we also pass index_col arguemnt which is "Respondent" here.
bd_df.head()
```

```python
# there might be some cases when we dont need to run the entire table.
# but we need to run a specific sql_query in order ti load in our data.
# to do this, we can use the method read_sql_query()
bd_db=pd.read_sql_query("SELECT * FROM bangla",engine,index_col="Respondent")
# here instead of the table name, we need to add the sql query.
bd_db.head()
```

```python
# that works too. we can use where in sql query to read data of certain conditions
# this is really useful when we are working with large database,
# but we need to work with specific data.

# we may have seen people load in data using an url instead of a specific file for some of the methods that we looked at before.
# we can do that.
# all we need to be sure that we are using correct method for the certain file format that on that url.
url_data=pd.read_json("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Flask_Blog/snippets/posts.json",engine)
url_data.head()
```
