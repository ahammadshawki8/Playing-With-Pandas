```python
import pandas as pd

df=pd.read_csv("stackoverflow/survey_results_public.csv",index_col="Respondent")
schema_df=pd.read_csv("stackoverflow/survey_results_schema.csv",index_col="Column")

pd.set_option("display.max_columns", 85)
pd.set_option("display.max_rows", 85)

df.head()
```

```python
# first lets look at some basics aggregations.
# aggregation basically means combining multiple pieces of data into a single result.
# for example, the functions that we usually use in statistics (mean,median,mode) ,
# these are aggregate functions.
# because they take multiple values and give us either mean or median or mode.

# ok, if we want to run a analysis on our dataframe here, we might ask a very common question-
# what is the typical salary of this survey respondant developers?
# that might be a good information to have if we are looking for a job and
# want to look what the current salary status of the developers in avarage.
# to do this we can grab the median salary of our dataframe.
```

```python
# first, lets look at the salary.
df["ConvertedComp"].head(15)
```

```python
# we can see the median slary just by running median() method on this series.
print(df["ConvertedComp"].median())
# this method will ignor NaN values.
```

```python
# but that doesn't gives us much information.
# since developers working in different countries are paid different amounts of salaries.
# so it would be nice if we could see the median salary broken down by the country.
# and we will look about that when we will learn grouping data.
# but first we need to learn few more concepts before we start grouping.
```

```python
# what if we run the median() function in our entire dataframe?
df.median()
```

```python
# when we do this it going to look through our dataframe 
# and find the columns which contains numerical values where it can grab a median value.
# some of this might not make sense to use with the median, but others might be pretty useful to us.
```

```python
# if we want to get a broad overview of our data and a statistical overview,
# then we can use the describe() method on our dataframe.
df.describe()
# this gives us count, mean, standerd diviation, minimum, 25%, 50%, 75% quantiles and maximum here.
# 50% is nothing but the median.
```

```python
# we might think why we have use median() method instead of mean() method which gives us the avarage?
# basically its because the mean() is affected to heavily by outlyers.
# its not a the good method to use because a few outlyers affect the avarage seriously.
# we can see that the mean is 127110 on avarage, where the median is 25777
# that gives us a unrealistic exceptation what is a typical developer salary is.
# because the largest salaries of our dataset are pulling up the avarage heavily.
# so in case of that, using median() instead mean() is recommanded. that is a better representation.
```

```python
# we can also run this descibe method in a single column as well.
df["ConvertedComp"].describe()
```

```python
# count() method just counts the not missing (not NaN) rows and 
# return the number of respondants (rows) who have answered the question.
df["ConvertedComp"].count()
```

```python
# some people think count() method counts up the individual values of every rows and 
# return the sum for each individual unique value.
# which is a wrong idea.

# if that what we want to do that whats value_counts() method do.
df["Hobbyist"].value_counts()
```

```python
# we can find some interesting information from our survey
# by using value_counts() method in different column.
# for example, there was a question about the social media used by the developers.
# to view this results we can access SocialMedia column.
# if we want to see the exact question, then we can find it from our schema dataframe
schema_df.loc["SocialMedia"]
```

```python
# lets access the SocialMedia column.
df["SocialMedia"]
```

```python
# we can also see the most popular social media by accessing value_counts() method in this column.
df["SocialMedia"].value_counts()
```

```python
# if we want to see this popularity by percentage instead of raw numbers,
# then we can pass in the normalize arguement to the value_counts() function and set that equal to True.
df["SocialMedia"].value_counts(normalize=True)
```

```python
# social media sites popularity varies a lot based on what country we are in.
# so how can we break up this results,
# so that we can see the most popular social media sites for each country?
# now in order to do this, we need to learn grouping our data.
```

```python
# first of all if we want to see specific result based on the country or any other column,
# then we have to group based on that soecific column.
# and we have the groupby() function for this.
# what does it actually mean?
# in pandas documentation, 
# it says that a groupby() operation involves some combinations of spliting up our object,
# applying different functions and combining our results.
# lets see each of those.
```

```python
# first just for a reference, lets print the value_counts() for countries.
df["Country"].value_counts()
```

```python
# now lets look how to use the groupby() function in our country column.
# first lets look at spliting the object.
country_grp = df.groupby(["Country"])
# this is going to be a list of columns that we going to group on.
# here we are only passing country column. But we have to use the third parenthesis[]
print(country_grp)
```

```python
# if we print that, what we are getting is-
# <pandas.core.groupby.generic.DataFrameGroupBy object at 0x000001BAC291C240>
# its a DataFrameGroupBy object.
# so what is this object and what exactly could we do with this?
# this object contains bunch of groups.
# and to better understand what this is lets look at a individual group that this dataframe has.
# since we grouped our rows by country, then we can grab a specific group by country name.
# to do this we can use get_group() function.
country_grp.get_group("United States")
# here we have to pass the column name as an arguement.
```

```python
# if we run that, then we can see that we are getting a dataframe back with some survey results.
# here all of the responses are that peoples who said that they are from United States.
# so thats what our dataframe groupby object consist of.
# this would be simmilar to running a filter on our original dataframe.
filt=(df["Country"]=="United States")
df.loc[filt]
```

```python
# so we can use filter for grabing the values based on a condition.
# But instead of just grabbing result from one country,
# group by splits up all of this respondants survey result by country name.
```

```python
# now that we have all of these information split up and grouped by country name,
# we can apply a function and bring that results back together.
# so what kind of function we like to apply.
# well we disscuss it before, maybe we want to see the mosst popular social media based on country.
# if we only want to see the popular social media for one country, then we have already seen how to do that.
df.loc[filt]["SocialMedia"].value_counts()
```

```python
# but if we run that on our dataframe groupby object, 
# then it will give us the result for all of those country groups.
country_grp["SocialMedia"].value_counts()
```

```python
# if we run this, it will return a series of most popular social platforms broken down by country. 
# but this is cut little bit early here.
# let us grab bigger chunk here to see how it looks like.
# we can do this by head() method.
country_grp["SocialMedia"].value_counts().head(50)
```

```python
# this is returning a series.
# this series have multiple indexes.
# but we haven't disscuss about multiple indexes yet.
# here country is the first index.
# we can grab this just like any other index that we grabbed earlier.
# if we want to grab values for bangladesh by index then we can use the loc[] indexer.
country_grp["SocialMedia"].value_counts().loc["Bangladesh"]
```

```python
# we may say that, here we are getting the same results that we get before.
# so why this is useful?
# this is useful because now we can see this results for any country
# without making a filter and running that for each and individual country of the world.
# for example, if we wanted to see the United Kingdom results,
# then without doing all those stuffs again and again we can just change an arguement and see them easily.
country_grp["SocialMedia"].value_counts(normalize=True).loc["United Kingdom"]
# if we want to print percentage without numbers then we can set normalize arguement to True.
```

```python
# now bringing that back about what we were disscussing at the beginning of the module.
# we can also use this for running more traditional aggregate functions like mean, median etc.
# before we look the median salary for entire survey, now lets break down them with countries instead.
# just like we see the results for sociial media column we can see that for ConvertedComp(salary) too.
country_grp["ConvertedComp"].median().head(20)
```

```python
# if we want to see the median result for germany then we can just-
country_grp["ConvertedComp"].median().loc["Germany"]
```

```python
# now maybe we are working on some analysis.
# there we not only have to group our data.
# but also we have to run multiple aggregrate function for our group.

# lets say we want to see both median and mean salary.
# for this we can use agg() method and
# pass in all the aggregate methods(this will be a str) in a list that we want to see.
country_grp["ConvertedComp"].agg(["mean","median"])
```

```python
# if we run this, then we can see that
# we are getting a dataframe of mean and median values for all countries.

# again if we want to see specific country result the we can use loc[] indexer.
country_grp["ConvertedComp"].agg(["mean","median"]).loc["Bangladesh"]
```

```python
# now depending on, what we are trying to do, we might run into some issues that we didnt expect.
# for example, lets say we want to figure out how many people of each country knows python.
# before we do this to our group, lets see how we can do this with a single country 
# using the filter approch that we do earlier.
# here we have to use the str.contains() method.
filt=(df["Country"]=="United States")
df.loc[filt]["LanguageWorkedWith"].str.contains("Python")
# this will return true if our respondant have worked with python.
# and will return false if he had not. 
```

```python
# now if we want to count all of the respondants number who worked with python.
# then we can use the sum() function.
# normally, people might think that sum() will only work for numeric values.
# but sum() will also work on booleans.
# it will count all Trues as 1 and all of the Falses as 0.
# so if we want to know the exact number of responders who know python,
# then we can simply add a .sum() in the end.
df.loc[filt]["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
# now before when we wanted to run a simmilar aggregation function on our dataframe gruopby object,
# we simply took the same approach on our group by object.
# for example,
# we might thing to do something like this to know how many people knew python from each country.
country_grp["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
# but if we run this here we can see that we are getting an attribute error.
# AttributeError: Cannot access attribute 'str' of 'SeriesGroupBy' objects, try using the 'apply' method

# here we are getting this error because this is no longer a series.
# instead, this is a dataframe groupby object.
# and we can use str methods in the groupby objects.
# we have to use apply() method instead and 
# have to specify a function that we want to run on every single series on this group.
country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
# here x is series.
```

```python
# Now seeing these numbers by itself isnt really that big help,
# if we are trying to get understanding of the percentage of python developers in each country.
# we can do that manually but that is too much pain.
# we want python and pandas to do this calculation for us.
```

```python
# instead, we are going to do that by the combination of all things 
# that we have already learned throughout the series.

country_respondants = df["Country"].value_counts()
country_uses_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
# here we have a variabls which represents a series of the total number of developers.
# we have another variabls which represents a series of the total number of python developers.
# now we need to combine this two.
# now we are actually going to us a method here that we havent disscussed yet.
# we probably should have mentioned that in that module when we appened dataframe.
# we can combine more than one series together using the pandas.concat() function.
```

```python
# lets create a dataframe by concatenating those two series.
py_df=pd.concat([country_respondants,country_uses_python],axis="columns",sort=False)
# here we need to pass the concateing serieses using a list as a arguement.
# we need also to pass another arguement which is axis and set that equals to "columns".
# we need to pass that because pandas actually try to concat serieses by rows in default.
# finally, we also need to pass sort arguemnt and set that equal to false.
# because if we havent done this, then we will have to get a warning.

# lets run this dataframe.
py_df
```

```python
# this is lot more useful for analyzing data.
# one thing that we might notice that,
# here we have columns which name does not relate to its data.
# so lets rename this so that they make more sense in the context that we wanted to do.
py_df.columns=["Total_Developer","Knows_Python"]
# we can also do that by-
# py_df.rename(columns={"Country":"Total_Developer","Total_Developer":"Knows_Python"})
```

```python
py_df
```

```python
# now lets create another column which represents the percentage of python developers.
py_df["Percentage"]= py_df["Knows_Python"]/py_df["Total_Developer"]*100
```

```python
py_df
```

```python
# now we can work these just like any other dataframe.
# lets say we want to sort these results by Percentage column.
py_df.sort_values(by="Percentage",ascending=False)
# we can use the inplace arguemnet to save those changes parmanenetly.
```

```python
# now that we have a dataframe we can see the info of a specific country.
py_df.loc["Bangladesh"]
```
